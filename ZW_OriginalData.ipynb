```python
import cbor
import sys
import gzip
import pprint
import os
import json
```

```python
pp = pprint.PrettyPrinter(indent=4)

infile = '/home/zengwei/DD/data/ebola-web/ebola-web-01/ebola-web-01-2016-16.cbor.gz'
if infile.endswith('.gz'):
    fp = gzip.open(infile, 'r')
else:
    fp = open(infile, 'r')

try:
    while (1):
        r = cbor.load(fp)
        pp.pprint(r)
except EOFError:
    pass
except:
    print >>sys.stderr, "Error reading", infile, sys.exc_info()[0]
```

```python
def count_records(infile):
    if infile.endswith('.gz'):
        fp = gzip.open(infile, 'r')
    else:
        fp = open(infile, 'r')

    Keys=[]
    try:
        while (1):
            thread = cbor.load(fp)
            Keys.append(thread.get(u'key'))
    except EOFError:
        return Keys
```

```python
FileInfo={}
dir = '/home/zengwei/DD/data/ebola-web/ebola-web-03'
for filename in os.listdir(dir): 
    if  filename.endswith('.gz') or filename.endswith('.cbor'):
        Keys = count_records(dir+'/'+filename)
        FileInfo[filename]=Keys
json.dump(FileInfo, open(dir +'-FileInfo.json', 'w'))
        
# count_records(infile)
```
